```python
#linear model to predict CO2emi from awareness
outcome_1,predictors_1 = patsy.dmatrices('CO2emi ~ Aware', df_numbers)
mod_1 = sm.OLS(outcome_1,predictors_1)
res_1 = mod_1.fit()

print(res_1.summary())
```

```python
#linear model to predict CO2emi from seriousness
outcome_1,predictors_1 = patsy.dmatrices('CO2emi ~ Serious', df_numbers)
mod_1 = sm.OLS(outcome_1,predictors_1)
res_1 = mod_1.fit()

print(res_1.summary())
```

```python
#linear model to predict awareness from seriousness
outcome_1,predictors_1 = patsy.dmatrices('Aware ~ Serious', df_numbers)
mod_1 = sm.OLS(outcome_1,predictors_1)
res_1 = mod_1.fit()

print(res_1.summary())
```

awareness predicts serioueness based on P value = 0.000 so is less < .01

awareness predicts CO2emi based on p value < .01

```python
#multivariate regression to predict CO2emi from both awareness and GDPperUS
outcome_2,predictors_2 = patsy.dmatrices('CO2emi ~ GDPperUS + Aware', df_numbers)
mod_2 = sm.OLS(outcome_2,predictors_2)
res_2 = mod_2.fit()

print(res_2.summary())
```

```python
#multivariate regression to predict CO2emi from both seriousness and GDPperUS
outcome_2,predictors_2 = patsy.dmatrices('CO2emi ~ GDPperUS + Serious', df_numbers)
mod_2 = sm.OLS(outcome_2,predictors_2)
res_2 = mod_2.fit()

print(res_2.summary())
```
